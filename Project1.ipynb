```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
data = pd.read_csv("time_series_2019-ncov-Confirmed.csv")
```

```python
data
```

```python
data=data.drop(['Province/State','Lat','Long'], axis=1)
df=data.groupby(["Country/Region"]).sum()
```

```python
df
```

```python
t = int(0.7*171)
print(t)
```

```python
 for col in df.columns: 
    if len(df[df[col] == 0])> t: 
        del df[col]
```

```python
df
```

```python
sns.barplot(data=df)
plt.xticks(rotation = 50)
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title("DATE WISE NUMBER OF CASES")
plt.show()
```
